```python
df_bezier=df_bezier.sort_values(by=["compression"],ascending=True)
```

```python
def get_best_of(compression,accuracy,skipC):
    i=0
    reduced_comp_list=[]
    reduced_acc_list=[]
    while i<len(compression):
        small_list=accuracy[i:i+skipC]
#         print(i,small_list)
        max_value = max(small_list)
        max_index = small_list.index(max_value)  
        reduced_comp_list.append(compression[i+max_index])
        reduced_acc_list.append(accuracy[i+max_index])      
#         print(reduced_comp_list,"\n",reduced_acc_list)
        i+=skipC
    return reduced_comp_list,reduced_acc_list
        

    
def get_max_of(compression,accuracy):
    i=1
    reduced_comp_list=[compression[0]]
    reduced_acc_list=[accuracy[0]]
    while i<len(compression):
#         print(i)
        if accuracy[i]>reduced_acc_list[-1]:
            reduced_acc_list.append(accuracy[i-1])
            reduced_comp_list.append(compression[i-1])
            
        else:
            i+=1            
            continue
        i+=1

    return reduced_comp_list,reduced_acc_list
        
        
        
    
```
