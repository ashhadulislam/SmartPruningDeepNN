```python
import pandas as pd
from matplotlib import pyplot as plt
import plotly.graph_objects as go
import numpy as np
import os
from scipy.signal import argrelextrema
from detecta import detect_peaks

from scipy import interpolate
import os
```

```python
files=os.listdir("results_sheet/noisy/")
print("TOtal files = ",len(files))
allowed_files_dic={}

for f in files:
    df=pd.read_csv("results_sheet/noisy/"+f)
    avg_acc=(df["val_1"].mean())
    allowed_files_dic[f]=avg_acc
    

allowed_files_dic_sorted={k: v for k, v in sorted(allowed_files_dic.items(), key=lambda item: item[1],reverse=True)}  
# print(allowed_files_dic_sorted)    


best_files=[]
for k,v in allowed_files_dic_sorted.items():
    if "03BezEvolMean" in k:
        print("GOt 3")
        best_files.append(k)
    elif "04BezActl" in k:
        print("GOt 4")        
        best_files.append(k)
    if len(best_files)==2:
        break

```
